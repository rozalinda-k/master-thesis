```python
final_list=[]
for index_primary_table,row_primary_table in cpa_test_gt.iterrows():
    list0=[]
    
    final_list_for_one_table=[]

    for i in range(5):
        
        x=[]
        x.append(row_primary_table[6][0][i][0][0])
        x.append(row_primary_table[6][0][i][1][1])
        list0.append(x)
        

    indexes=[]
    for value in list0:
        indexes.append(value[0])

    
    for column in row_primary_table[6]:
        list1=[]

        for column_index in column:
            
            if(column_index[0][1] != 0):
                column_label="Column{}".format(2)
                if column_index[0][0] in indexes:
                    x=[]
                    x.append(column_index[0][0])
                    x.append(column_index[1][1])
                    list1.append(x)
                else:
                    x=[]
                    x.append(" ")
                    x.append(" ")
                    list1.append(x)

        Column_labels="Column 1 || Column {} ||\n ".format(2)
        
        combination_values=""
        final_combination="{}{}"
        for index_list0,value0 in enumerate(list0):
            for index_list1,value1 in enumerate(list1):

                if value0[0]== value1[0]:
                    combination_values=combination_values+value0[1]+" || "+value1[1]+" ||\n "


        final_combination=final_combination.format(Column_labels,combination_values)

        
        if len(final_combination)>30:
            unchecked_labels=row_primary_table[7]
            labels_in_all_labels=[]
            for label in unchecked_labels:
                check_label=False
                for inner_label in label:
                    if inner_label in all_labels:
                  
                        check_label=True
                    else:
                        label.remove(inner_label)
                    if(len(label)!=0):
                        labels_in_all_labels.append(label)
                    
            if  check_label==True:
                final_list_for_one_table.append(row_primary_table[0])
                final_list_for_one_table.append(final_combination)
                final_list_for_one_table.append(labels_in_all_labels)

    if len(final_list_for_one_table)>1:
        final_list.append(final_list_for_one_table)
        
test_examples = final_list
```
